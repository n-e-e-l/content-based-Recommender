```python
#getting the details
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Get first 3 cast,keyword and genre details if they are present
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
import numpy as np

# Extracting director, cast, genres and keywords features that are in a suitable form.
df_movies_metadata['director'] = df_movies_metadata['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df_movies_metadata[feature] = df_movies_metadata[feature].apply(get_list)
```

```python
# Function to clean the newly structred data 
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Tag formation
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df_movies_metadata[feature] = df_movies_metadata[feature].apply(clean_data)
```

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw = stopwords.words('english')
```

```python
#sw = stopwords.words('english')
def stopwords(text):
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    return " ".join(text)

df_movies_metadata['combine_data'] = df_movies_metadata['combine_data'].apply(stopwords)
```

```python
stemmer = SnowballStemmer("english")
def stemming(text):    
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text)
df_movies_metadata['combine_data'] = df_movies_metadata['combine_data'].apply(stemming)
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
# Apply to the DF series
df_movies_metadata['combine_data'] = df_movies_metadata['combine_data'].apply(remove_punctuations)
```

```python
df_movies_metadata['combine_data'] = df_movies_metadata['combine_data'].str.replace('\d+', '')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vector = cv.fit_transform(df_movies_metadata['combine_data']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = df_movies_metadata[df_movies_metadata['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df_movies_metadata.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
